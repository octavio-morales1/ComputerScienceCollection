 1 Kernel Ridge Regression [45 pts]
 a. Load the provided data sets. The train data set will be used to train the model.

Importing Modules:

```python
import pandas as pd
import numpy as np
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error

# below is to allow users to avoid seeing the warnings
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=RuntimeWarning)
warnings.filterwarnings("ignore", category=ConvergenceWarning)

```

```python
print('Krenel Ridge Train:')
dTrain= pd.read_csv('kernel_ridge_train.csv')
print(dTrain)
```

```python
print('Krenel Ridge Test:')
dTest = pd.read_csv('kernel_ridge_test.csv')
print(dTest)
xT= dTrain.iloc[:, :-1].values
yT= dTrain.iloc[:, -1].values
testX= dTest.iloc[:, :-1].values
testY= dTest.iloc[:, -1].values
```

b. [10 pts] Train the NumPy implemented kernel ridge regression model with the polynomial kernel function, k = ⟨x,z⟩d. Find the best model with the lowest MSE value as the degree, d, increases from 2 to 11.

```python
pMSE= float('inf')
pDEGREE= 0
for degree in range(2, 12):
    temp= KernelRidge(kernel='polynomial', degree=degree, alpha=1.0)
    temp.fit(xT, yT)
    tempY= temp.predict(testX)
    mse= mean_squared_error(testY, tempY)
    if (mse< pMSE):
        pMSE= mse
        pDEGREE= degree
print(f"Best Polynomial Degree: {pDEGREE}, MSE: {pMSE}")
```

c. [10 pts] Write a code that constructs the kernel matrix with RBF kernel function, k(x,z) = exp {−γ|x−z|^2} .

```python
rbfMSE= float('inf')
rbfG= 0
gamma_values= np.logspace(-2, 2, 10)
for gamma in gamma_values:
    temp= KernelRidge(kernel='rbf', gamma=gamma, alpha=1.0)
    temp.fit(xT, yT)
    tempY= temp.predict(testX)
    mse= mean_squared_error(testY, tempY)
    if(mse < rbfMSE):
        rbfMSE= mse
        rbfG= gamma
print(f"Best RBF Gamma: {rbfG}, MSE: {rbfMSE}")
```

d. [15 pts] Construct a kernel function, the product poly-kernel and rbf-kernel functions, and train the kernel ridge regression model. Find the best model with the combinations of d and γ values used in b and c. There must be a total of 100 combinations.

```python
pMSE= float('inf')
bDEGREE= 0
for degree in range(2, 12):
    kr= KernelRidge(alpha=1.0, kernel='polynomial', degree=degree)
    kr.fit(xT, yT)
    tempY= kr.predict(testX)
    mse= mean_squared_error(testY, tempY)
    if(mse< pMSE):
        pMSE= mse
        bDEGREE= degree
print(f"Polynomial Kernel: Best Degree = {bDEGREE} with MSE = {pMSE}")

rbfMSE= float('inf')
bestG= 0
for g in np.logspace(-2, 2, 10):
    kr= KernelRidge(alpha=1.0, kernel='rbf', gamma=g)
    kr.fit(xT, yT)
    tempY= kr.predict(testX)
    mse= mean_squared_error(testY, tempY)
    if(mse < rbfMSE):
        rbfMSE= mse
        bestG= g
print(f"RBF Kernel: Best Gamma = {bestG} with MSE = {rbfMSE}")
```

e. [5 pts] Using the best models found in b, c, and d, pick the best model. Explain your choice.

```python
temp= 0
if(pMSE< rbfMSE):
    print(f"The best model is Polynomial Kernel Ridge Regression with degree {bDEGREE}, having an MSE of {pMSE}.")
    temp= KernelRidge(alpha=1.0, kernel='polynomial', degree=bDEGREE)
else:
    print(f"The best model is RBF Kernel Ridge Regression with gamma {bestG}, having an MSE of {rbfMSE}.")
    temp= KernelRidge(alpha=1.0, kernel='rbf', gamma=bestG)
temp.fit(xT, yT)
```

f. [5 pts] Generalize the best train model and predict the target.

```python
yBest = temp.predict(testX)
bestMSE = mean_squared_error(testY, yBest)
print(f"MSE of the Best Model on Test Data: {bestMSE}")
```
